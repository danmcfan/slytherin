```python
%matplotlib widget

import numpy as np
from matplotlib import pyplot as plt


def plot_graph(
    x_range_min,
    x_range_max,
    y_range_min,
    y_range_max,
    function,
    x_label,
    y_label,
    z_label,
    interval=1,
    projection_3d=False,
):
    x_range = np.arange(x_range_min, x_range_max, interval)
    y_range = np.arange(y_range_min, y_range_max, interval)

    x_grid, y_grid = np.meshgrid(x_range, y_range)

    z_grid = function(x_grid, y_grid)

    plt.figure(figsize=(5, 5))

    if projection_3d:
        ax = plt.axes(projection="3d")
    else:
        ax = plt.axes()

    ax.contour(x_grid, y_grid, z_grid, 100, cmap="binary")

    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)

    if projection_3d:
        ax.set_zlabel(z_label)
        ax.view_init(50, 130)
    else:
        ax.set_title(z_label)

    plt.show()

    max_z = np.max(z_grid)

    x_index, y_index = np.where(z_grid == max_z)

    x = x_range[x_index][0]
    y = y_range[y_index][0]

    print(
        f"Maximum {z_label} is {round(max_z, 2)} at {round(x, 2)} {x_label} and {round(y, 2)} {y_label}"
    )

    min_z = np.min(z_grid)

    x_index, y_index = np.where(z_grid == min_z)

    x = x_range[x_index][0]
    y = y_range[y_index][0]

    print(
        f"Minimum {z_label} is {round(min_z, 2)} at {round(x, 2)} {x_label} and {round(y, 2)} {y_label}"
    )
```

```python

from config.happiness import (
    FUNCTION,
    X_RANGE_MIN,
    X_RANGE_MAX,
    Y_RANGE_MIN,
    Y_RANGE_MAX,
    X_LABEL,
    Y_LABEL,
    Z_LABEL,
)

plot_graph(X_RANGE_MIN, X_RANGE_MAX, Y_RANGE_MIN, Y_RANGE_MAX, FUNCTION, X_LABEL, Y_LABEL, Z_LABEL)
```
