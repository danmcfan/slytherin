```python
%matplotlib widget

import numpy as np
from matplotlib import pyplot as plt


from config.happiness import (
    FUNCTION,
    X_RANGE_MIN,
    X_RANGE_MAX,
    Y_RANGE_MIN,
    Y_RANGE_MAX,
    X_LABEL,
    Y_LABEL,
    Z_LABEL,
)

PROJECTION_3D = False
INTERVAL = 0.1

x_range = np.arange(X_RANGE_MIN, X_RANGE_MAX, INTERVAL)
y_range = np.arange(Y_RANGE_MIN, Y_RANGE_MAX, INTERVAL)

x_grid, y_grid = np.meshgrid(x_range, y_range)

z_grid = FUNCTION(x_grid, y_grid)

fig = plt.figure(figsize=(5, 5))

if PROJECTION_3D:
    ax = plt.axes(projection="3d")
else:
    ax = plt.axes()

ax.contour(x_grid, y_grid, z_grid, 100, cmap="binary")

ax.set_xlabel(X_LABEL)
ax.set_ylabel(Y_LABEL)

if PROJECTION_3D:
    ax.set_zlabel(Z_LABEL)
    ax.view_init(50, 130)
else:
    ax.set_title(Z_LABEL)

plt.show()

max_z = np.max(z_grid)

x_index, y_index = np.where(z_grid == max_z)

x = x_range[x_index][0]
y = y_range[y_index][0]

print(
    f"Maximum {Z_LABEL} is {round(max_z, 2)} at {round(x, 2)} {X_LABEL} and {round(y, 2)} {Y_LABEL}"
)

min_z = np.min(z_grid)

x_index, y_index = np.where(z_grid == min_z)

x = x_range[x_index][0]
y = y_range[y_index][0]

print(
    f"Minimum {Z_LABEL} is {round(min_z, 2)} at {round(x, 2)} {X_LABEL} and {round(y, 2)} {Y_LABEL}"
)
```
